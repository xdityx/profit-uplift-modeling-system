# Result Analysis and Targeting Policy Insights

This notebook summarizes the experimental results of different targeting strategies.

The objective is to analyze how predictive and causal models influence marketing profit and decision policies.

```python
import pandas as pd

results = pd.DataFrame({
"Strategy":["Random","Predictive","T-Learner","X-Learner"],
"Profit":[32040,150000,117780,77640]
})

results
```

```python
import matplotlib.pyplot as plt

plt.bar(results["Strategy"],results["Profit"])

plt.title("Profit Comparison Across Targeting Strategies")
plt.ylabel("Profit")
plt.xlabel("Targeting Strategy")

plt.show()
```

## Result Interpretation

The predictive targeting strategy achieved the highest profit in this simulation.

This occurs because baseline conversion probability dominates treatment effect magnitude in the generated dataset.

Customers with high predicted probability are also highly likely to convert when targeted.

Therefore predictive ranking captures most profitable customers even without modeling treatment effect.

## When Uplift Modeling Provides Advantage

Uplift models outperform predictive models when:

1. Treatment effect heterogeneity is large
2. Baseline conversion probability is weakly correlated with treatment effect
3. Some customers respond negatively to treatment
4. Campaign incentives significantly change behavior

These conditions commonly appear in targeted marketing campaigns and personalized recommendation systems.

## Key Insight

Predictive modeling optimizes:

P(Y=1 | X)

Uplift modeling optimizes:

E[Y(1) − Y(0) | X]

This difference reflects the distinction between **correlation-based prediction** and **causal decision-making**.
